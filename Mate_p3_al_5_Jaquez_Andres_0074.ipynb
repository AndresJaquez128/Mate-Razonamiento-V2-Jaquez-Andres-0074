###Problema 3
Paso 1: Análisis y resolución paso a pasoProblema 3: Un número al que se le resta siete da 18Definir la variable: Sea $x$ el número buscado.Plantear la ecuación: "Un número ($x$) al que se le resta siete ($- 7$) da 18 ($= 18$)".$$x - 7 = 18$$Despejar: El 7 que está restando pasa al otro lado sumando.$$x = 18 + 7$$Resultado: $x = 25$

```python
# Problema 3: x - 7 = 18
#resultado = 18
#resta = 7
resultado = int(input("Ingrese el resultado: "))
resta = int(input("Ingrese la resta: "))
# Al despejar, la resta pasa como suma
x = resultado + resta
print(f"El número del problema 3 es: {x}")
```

##Problema 4:
 El triple de un número más cuatro es 31Definir la variable: Sea $x$ el número.Plantear la ecuación: "El triple de un número ($3x$) más cuatro ($+ 4$) es 31 ($= 31$)".$$3x + 4 = 31$$Despejar: Primero movemos el 4 (pasa restando).$$3x = 31 - 4$$$$3x = 27$$Resolver: El 3 que multiplica pasa dividiendo.$$x = \frac{27}{3}$$Resultado: $x = 9$

```python
# Problema 4: 3x + 4 = 31
#total = 31
#suma = 4
#multiplo = 3
total = int(input("Ingrese el total: "))
suma = int(input("Ingrese la suma: "))
multiplo = int(input("Ingrese el multiplo: "))
# Primero restamos el valor independiente y luego dividimos
x = (total - suma) / multiplo
print(f"El número del problema 4 es: {int(x)}")
```

## Problema 5:
 Terreno rectangular (Largo = 5 veces el ancho, P = 50m)Definir las variables:Ancho: $x$Largo: $5x$Plantear la ecuación: El perímetro ($P$) de un rectángulo es la suma de sus cuatro lados: $P = 2(largo) + 2(ancho)$.$$2(5x) + 2(x) = 50$$Simplificar:$$10x + 2x = 50$$$$12x = 50$$Despejar:$$x = \frac{50}{12}$$Resultado: * Ancho: $4.166...$ metros.Largo: $5 \times 4.166 = 20.833...$ metros.

```python
# Problema 5: 12x = 50 (Suma de todos los lados)
#perimetro_total = 50
#total_unidades_x = 12 # 2 lados de 5x + 2 lados de 1x
perimetro_total = int(input("Ingrese el perímetro total: "))
total_unidades_x = int(input("Ingrese el total de unidades en x: "))
ancho = perimetro_total / total_unidades_x
largo = ancho * 5

print(f"Ancho del terreno: {round(ancho, 2)}m")
print(f"Largo del terreno: {round(largo, 2)}m")
```
